```python
from matplotlib.pyplot import plot
import pandas as pd
import numpy as np
```

```python
def activation(z,A):
    A=[]
    for i in range(0,4):
        c=z[0,i]
        a=1/(1+np.exp(-c))
        A.append(a)
    return A

#initialisation

Y=np.mat([0,1,1,0])   #sortie voulue 
W1=np.mat('[-1.5, 1., 1. ; -0.5, 1., 1.]')  #matrice biais et le poids
MX=np.mat([[1,1,1,1], [0,0,1,1], [0,1,0,1]]) #entrées X1 et X2+ligne des 1 pour le biais
print("MX=", MX)

```

```python
#fonction de forward
def forward(W1,MX):
    M3 = W1.dot(MX)  #produit matriciel pour obtenir z1 et z2
    print("produit matriciel = ", M3)
    
    z1=M3[0]  #les z1 correspondent à la 1ere ligen de M3
    z2=M3[1]  #les z2 correspondent à la 2eme ligne de M3
    print("z1= ", z1, "z2 = ", z2)
    
    #activation de la 2eme couche 
    a1=[]
    a2=[]
    a1=activation(z1,a1)  #pren la 1ere ligne de M3 et on l'active pour obtenir a1
    a2=activation(z1,a2)  #pren la 2ere ligne de M3 et on l'active pour obtenir a2
    print("a1 = ", a1, "a2 = ", a2)
    
    #on crée une matrice nulle de 2 lignes pour y injecter les a1 et les a2
    MZ=np.zeros((2, 4))
    MZ[0]=a1
    MZ[1]=a2
    
    
    W2 = np.mat('[-0.5, -1., 1.]') #le biais et le poids
    C=[1,1,1,1]  #ligne de 1 pour multiplier les biais
    MZ = np.row_stack((C, MZ)) #ajoute de C dans MZ
    Z3=np.dot(W2, MZ)
    print("MZ=", MZ)
    print("Z3 = ", Z3)
    
    #activation de la 3eme couche 
    
    a3=[]
    a3=activation(Z3,a3)
    print("a3 = ", a3)
    

    return a1,a2,Z3
     
```

```python
#fonction loss
alpha = 0.01

def derive_a3_z3(a3,z3):
    D=(np.exp(-z3)/(1+np.exp(-z3))**2)
    return D

def derive_L_a3(a3,Y):
    d=(1/2)*(a3-Y)
    return d

def derive_z3_w(a, i):
    ax = a[i]
    return ax

#Fonction cout

def cout(a3,Y):
    l=0
    for i in range(0,4):
        l += 1/4 * ((a3[i]-Y[:,i])**2)
    print("l = ", l)
    return float(l)




ML = []
N = []
for i in range(1000):
    a1, a2, Z3 = forward(W1, MX)  #appele  la fonction forward pour pouvoir utiliser les memes matrices et var
    l=cout(a3,Y)
    ML.append(l)
    N.append(i)
    if l> alpha:
        for i in range(0,4):
            for k in range(1,3):
                z3=Z3[:,i]
                x=a3[i]
                y=Y[:,i]
                 #Prend la i eme colonne de Y
                d1=float(derive_a3_z3(x,z3))
                d2=float(derive_L_a3(x,y))
                d3=derive_z3_w(a1,i)
                L=d1*d2*d3
                if L!=0 and i==1 :
                    W1 [0,k]=W1 [0,k] - alpha*L
                    print ("W1 = ", W1)
                    a1,a3,Z3 = forward(W1, MX)
                elif L!=0 and i==2 :
                    W1 [1, k]=W1 [1,k] - alpha*L
                    print("W1 = ", W1)
                else :
                    print(W1[k-1, k])
    else :
        print("W1new = ", W1)
        
plot(N, ML)



    
    
```
