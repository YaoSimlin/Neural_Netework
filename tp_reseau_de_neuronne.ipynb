```python
# Sélectionner la colonne cible
pred = df[["Like"]]  # Cible

# Sélectionner les colonnes explicatives en enlevant "ID", "Name", et "Like"
feat = df.drop(columns=["ID", "Name", "Like"])  # Variables explicatives

# Définir X et Y
X = feat  # Variables explicatives
Y = pred  # Cible
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size=0.25, random_state=0)
print("les données d'entrainement Xi : ", x_train.shape)
print("les données d'entrainement Y:", y_train.shape)
print("les données de test Xi:", x_test.shape)
print("les données de test Y: ", y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout


network=Sequential()
network.add(Dense(1, kernel_initializer='uniform', activation='relu', bias_initializer='zero', input_shape=(31,)))
network.add(Dense(1, kernel_initializer='uniform', activation='relu'))
#1 utilise  un noeud entre la couche cachée et celle de la couche d entree
```

```python
network.summary()
```

```python
network.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = network.fit(x_train, y_train, batch_size=32, epochs=100)
```

```python
x_new=x_test[:2] # Selection of the first line of x_test matrix
y_nex=y_test[:2]
x_new
```

```python
#TESTING THE MODEL
```

```python
loss,score = network.evaluate(x_test,y_test)
print('loss :',loss)
print('score :',score)
```
